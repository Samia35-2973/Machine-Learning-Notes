```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('carprices.csv')
df
```

```python
dummies = pd.get_dummies(df[['Car Model']])
dummies
```

```python
df = pd.concat([df, dummies], axis='columns')
df
```

```python
df = df.drop(['Car Model', 'Car Model_Audi A5'], axis='columns')
df
```

```python
y = df[['Sell Price($)']]
y
```

```python
x = df.drop(['Sell Price($)'], axis='columns')
x
```

```python
model = linear_model.LinearRegression()
model.fit(x,y)
```

```python
# Predict price of a mercedez benz that is 4 yr old with mileage 45000
model.predict([[45000, 4, 0, 1]])
```

```python
# Predict price of a BMW X5 that is 7 yr old with mileage 86000
model.predict([[86000, 7, 1, 0]])
```

```python
model.score(x,y)
```

```python
le = LabelEncoder()
```

```python
d = pd.read_csv('carprices.csv')
dle = d
dle['Car Model'] = le.fit_transform(dle['Car Model'])
dle
```

```python
X = dle[['Car Model', 'Mileage', 'Age(yrs)']].values
X
```

```python
Y = dle[['Sell Price($)']].values
Y
```

```python
ohe = OneHotEncoder()
ct = ColumnTransformer([('Car Model', ohe, [0])], remainder='passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]
X
```

```python
model.fit(X, Y)
```

```python
# Predict price of a mercedez benz that is 4 yr old with mileage 45000
model.predict([[0, 1, 45000, 4]])
```

```python
# Predict price of a BMW X5 that is 7 yr old with mileage 86000
model.predict([[1, 0, 86000, 7]])
```

```python
model.score(X,Y)
```
