## [Titanic Kaggle Competition](https://www.kaggle.com/competitions/titanic/overview)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn import tree
import math
```

```python
df = pd.read_csv('train.csv')
df
```

```python
d = df[['Pclass', 'Sex', 'Age', 'Fare', 'Survived']]
d
```

```python
# to check if any row of this column has nan value
d['Pclass'].isna().any()
```

```python
d['Sex'].isna().any()
```

```python
d['Age'].isna().any()
```

```python
d['Fare'].isna().any()
```

```python
d['Survived'].isna().any()
```

```python
# since age has nan value so we will take mean value of it
mean_age = math.floor(d['Age'].mean())
mean_age
```

```python
d['Age'] = d['Age'].fillna(mean_age)
d
```

```python
d['Age'].isna().any()
```

```python
le = LabelEncoder()
```

```python
d['Sex'] = le.fit_transform(d['Sex'])
d
```

```python
x_train = d[['Pclass', 'Sex', 'Age', 'Fare']].values
x_train
```

```python
ohe = OneHotEncoder()
ct = ColumnTransformer([('Pclass', ohe, [0])], remainder='passthrough')
```

```python
x_train = ct.fit_transform(x_train)
```

```python
x_train = x_train[:,1:]
x_train
```

```python
y_train = d[['Survived']].values
y_train
```

```python
model = tree.DecisionTreeClassifier(criterion='entropy')
```

```python
model.fit(x_train, y_train)
```

```python
test_df = pd.read_csv('test.csv')
test_df
```

```python
test_df['Pclass'].isna().any()
```

```python
test_df['Sex'].isna().any()
```

```python
test_df['Age'].isna().any()
```

```python
test_df['Fare'].isna().any()
```

```python
test_df['Age'] = test_df['Age'].fillna(math.floor(test_df['Age'].mean()))
test_df['Age'].isna().any()
```

```python
test_df['Fare'] = test_df['Fare'].fillna(math.floor(test_df['Fare'].mean()))
test_df['Fare'].isna().any()
```

```python
test_df['Sex'] = le.fit_transform(test_df['Sex'])
test_df
```

```python
ct2 = ColumnTransformer([('Pclass', ohe, [0])], remainder='passthrough')
```

```python
x_test = test_df[['Pclass', 'Sex', 'Age', 'Fare']].values
x_test
```

```python
x_test = ct2.fit_transform(x_test)
x_test = x_test[:, 1:]
x_test
```

```python
y_predicted = model.predict(x_test)
```

```python
predicted_df = pd.DataFrame(test_df['PassengerId'], columns=['PassengerId'])
predicted_df
```

```python
predicted_df['Survived'] = y_predicted
predicted_df
```
