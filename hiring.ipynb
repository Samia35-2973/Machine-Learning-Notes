```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
from sklearn import linear_model
from word2number import w2n # for converting word to number
```

```python
df = pd.read_csv('hiring.csv')
df
```

```python
df.isna()
```

```python
# Since experience has also string value so null values can be stated as 'zero'
df.experience = df.experience.fillna('zero')
df
```

```python
# convert the words to number
df.experience = df.experience.apply(w2n.word_to_num)
```

```python
# Since there are lot of data mean is prefered

df[['test_score(out of 10)']] = df[['test_score(out of 10)']].fillna(math.floor(df[['test_score(out of 10)']].mean()))
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']], df[['salary($)']])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2, 9, 6]])
```
