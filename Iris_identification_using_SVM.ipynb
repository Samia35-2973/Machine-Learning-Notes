```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
iris = load_iris()
```

We will load the iris data to a dataframe for a better view

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df
```

Append the target column on the dataframe

```python
df['Target'] = iris.target
df
```

Now If we explore dataframe by target there are 3 values: 0, 1, 2 indicating iris types.

```python
df[df.Target==0]
```

```python
df[df.Target==1]
```

```python
df[df.Target==2]
```

Here We can conclude that 0-49 --> 1, 50-99 -->2, 100-149 -->3

From one column we will generate another column. We will generate flower names from target. For that we will apply and use lambda function for transformation. Here lambda will return for each value x of target, it will put target names

```python
df['Flower_names'] = df.Target.apply(lambda x: iris.target_names[x])
df
```

Now we will visualize the data

```python
df1 = df[df.Target==0]
df2 = df[df.Target==1]
df3 = df[df.Target==2]
```

```python
%matplotlib inline
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='green', marker='+')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color='red', marker='+')
plt.scatter(df3['sepal length (cm)'], df3['sepal width (cm)'], color='blue', marker='+')
```

```python
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='yellow', marker='+')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'], color='black', marker='+')
plt.scatter(df3['petal length (cm)'], df3['petal width (cm)'], color='pink', marker='+')
```

We are going to use all 4 features and the classification is also going to be in all 3 species

```python
x = df[['sepal length (cm)',	'sepal width (cm)',	'petal length (cm)',	'petal width (cm)']]
x
```

```python
y = df[['Target']]
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
model = SVC() # here we can pass different c-regularization, gamma and different kernal
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
