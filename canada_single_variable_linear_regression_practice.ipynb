```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter(df[['year']], df[['per capita income (US$)']], color='red', marker='*')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df[['per capita income (US$)']])
```

```python
reg.predict([[2011]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
828.46507522 * 2011 + -1632210.75785546
```

```python
d = pd.read_csv('years.csv')
d
```

```python
p = reg.predict(d)
d['per capita income (US$)'] = p
```

```python
d.to_csv('submission.csv')
```

```python
%matplotlib inline
plt.xlabel('years')
plt.ylabel('per capita income (US$)')
plt.scatter(df[['year']], df[['per capita income (US$)']], color='red', marker='+')
plt.scatter(df[['year']], reg.predict(df[['year']]), color='blue', marker='*')
```
