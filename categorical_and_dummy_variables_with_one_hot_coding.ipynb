```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
# Create Dummy Variable columns
# Pandas has the get dummy method that only return dummy variable columns
dummies = pd.get_dummies(df.town)
dummies
```

```python
# Concatenate dummy dataframe with original dataframe. Pandas has the concat method which takes 2 dataframes tha to be merged and in which axis to merge
merged = pd.concat([df, dummies], axis='columns')
merged
```

```python
# Now we need to drop the town column as it's no longer needed.
# We also need to drop one of the dummy variable columns. It's important as it avoids dummy variable trap
# on data frame call drop method indicating columns and axis
final = merged.drop(['town', 'west windsor'], axis='columns')
final
# Sklearn drops and avoid trap automatically in linear regression but it's a good practice to drop own
```

```python
# drop the column of price
y = final.price
y
```

```python
x = final.drop('price', axis='columns')
x
```

```python
# Now build model as usual
model = linear_model.LinearRegression()
model.fit(x, y)
```

```python
model.predict([[2800, 0, 1]])
```

```python
# If want to know how accurate model is apply score method and supply x and y.
# It will take all x and y, and compare with predicted value with some formula and generate score
model.score(x, y)
```

```python
# Create dummy variables
from sklearn.preprocessing import OneHotEncoder
# Create one hot encoder object and pass the 0th column on that x as categorical feature
# ohe = OneHotEncoder(categorical_features = [0]) # Here assign the index of column no as category to transform that column only
# Now this don't work in this version

ohe = OneHotEncoder()
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', ohe, [0])], remainder = 'passthrough')
```

```python
# Fit and transform
# If we pass X data frame it will assume all the columns as categorical columns and transform and convert to array
X = ct.fit_transform(X)
X
```

```python
# Avoid Dummy variable trap
X = X[:,1:] # Take all rows on 1st argument,take 1 to all columns on 2nd argument
X
```

```python
model.fit(X,Y)
```

```python
model.predict([[0,1,3400]])
```
