```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import linear_model
import math
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
# In the training data sometimes datapoints go missing from a cell or it can be seen as Nan
# Handle missing data points and there is a linear relation in this data points

# price = m1* area + m2 * bedrooms + m3 * age + b here area, bedrooms and age is the independent variables of features
# m1, m2, m3 are coefficients
```

```python
# Check in which column there are Nan values
# for that column we have to replace all Nan values with the mean or median of the full column
# Take mean when there are lots of data otherwise median will be enough

# Check if Nan cells are present it will return true
df.isna()
```

```python
# Take panda series of the table.column and then call median function
df.bedrooms.median()

# to keep it integer we can round down and for that use floor funtion
median_bedrooms = math.floor(df.bedrooms.median())
```

```python
# to fill column's na value use fillna available on panda series
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
df

# till now it was data preprocessing or clean data. It is always necessary to clean data before using a ml model
```

```python
# Train model

reg = linear_model.LinearRegression()
reg.fit(df[['area', 'bedrooms', 'age']], df.price)
```

```python
reg.coef_ # It will give an array containing all m values
```

```python
reg.intercept_
```

```python
reg.predict([[3000, 3, 40]])
```
